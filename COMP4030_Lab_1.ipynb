```python
# COMBINE tables with PANDAS

# first divided tables, first 15 rows
df_1 = pd.read_csv('countries.csv', nrows=15)
df_2 = pd.read_csv('countries.csv', skiprows=range(1,16))
```

```python
print(df_1.shape)
df_1
```

```python
print(df_2.shape)
df_2
```

```python
# Combine tables VERTICALLY -- adding rows
df = pd.concat([df_1, df_2], axis=0).reset_index(drop=True)
df
```

```python
# Combine tables HORIZONTALLY -- adding columns

# Combine df_A and df_B
df_A = df[['COUNTRY', 'GDP']].sort_values(by=['GDP']).iloc[2:]
df_A
```

```python
df_B = df_2[['COUNTRY', 'AREA', 'POP']]
df_B
```

```python
pd.merge(df_A, df_B,  on = 'COUNTRY', how='inner')
```

```python
# Final Task 1 - new column density of population of every country in people per square meter
df_countries['DENSITY'] = df_countries['POP'] / df_countries['AREA_SQUARE_KM']
df_countries
```

```python
# Final Task 2 - What continent has the countries with largest density of population on average
df_sort_3 = df_countries.sort_values(by = 'DENSITY', ascending = False)
df_sort_4 = df_sort_3.groupby(['CONT']).mean().sort_values(by='DENSITY', ascending=False)
df_sort_4
```
